```python
# %load zomato_selenium.py
import time
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException

restorant_data = []

option = webdriver.ChromeOptions()
option.add_argument(" -- incognito")

driver = webdriver.Chrome("./chromedriver", chrome_options=option)

for pg in [1,2,3,4]:
    driver.get("https://www.zomato.com/bhopal/bhel-restaurants?page="+str(pg))
    time.sleep(5)

    wait = WebDriverWait(driver, 30)
    body = wait.until(EC.presence_of_element_located((By.XPATH,"/html/body")))

    name = driver.find_elements_by_class_name("search-snippet-card")

    for nm in name:
        try:
            article = nm.find_element_by_tag_name('article')
        except:
            continue
        ## result_title
        try:
            result_title = article.find_element_by_class_name('result-title')
            result_title_text = result_title.text
        except:
            result_title_text = ""
        ## subzone
        try:
            search_result_subzone = article.find_element_by_class_name('search_result_subzone')
            subzone = search_result_subzone.text
        except:
            subzone = ""
        try:
            search_result_rating = article.find_element_by_class_name("search_result_rating")
            rating_popup = search_result_rating.find_element_by_class_name("rating-popup")
            rating = rating_popup.text
            try:
                voting_span = search_result_rating.find_element_by_tag_name("span")
                voting = voting_span.text
            except:
                voting = ""
        except:
            rating=""
            voting=""
        try:
            search_result_address = article.find_element_by_class_name("search-result-address")
            address = search_result_address.text
        except:
            address = ""
        try:
            search_page_text_div = article.find_element_by_css_selector("div.search-page-text").find_elements_by_tag_name('div')
            cuisines = tryexcept(search_page_text_div[0].text.split('\n')[1])
            cost= tryexcept(search_page_text_div[1].text.split('\n')[1])
            hours = tryexcept(search_page_text_div[2].text.split('\n')[1]).split('(')
            times = hours[0].strip()
            weeks = hours[1].strip().replace(')','')
        except:
            cuisines = cost = times = weeks = ""
        restorant_data.append([result_title_text, subzone, rating, voting, address, cuisines, cost, times, weeks])

import pandas as pd

df = pd.DataFrame(restorant_data, columns=['title', 'subzone', 'rating', 'voting', 'address', 'cuisines', 'cost', 'times', 'weeks'])
df
#df.to_csv('restorant.csv', sep='\t', encoding='utf-8')
```

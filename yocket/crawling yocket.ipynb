```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

temp=[]


for i in range(1,10):
    page = requests.get("https://yocket.in/applications-admits-rejects/1282-georgia-state-university/{}".format(i))

    soup = BeautifulSoup(page.content, 'html.parser')



    name_containers = soup.find_all('div', class_ = 'col-sm-6')
    for i in name_containers:
        k =(i.div.text)
        t=[i for i in k.strip().split("\n") if len(i) is not 0]
        temp.append(t)



pandas= pd.DataFrame(temp)
pandas.columns=["name","university","admit season","interested","Exam","Score","Eng Exam","Eng Score","Education","Education score","Work Exp","Exp in Months"]



```
